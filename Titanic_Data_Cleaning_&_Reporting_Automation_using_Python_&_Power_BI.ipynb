```python
import pandas as pd
df=pd.read_csv("Titanic-Dataset.csv")
```

```python
print("\n500 rows")
df.head(891)

```

```python
print("\nDataset Info")
print(df.info())

print("\nDataset Shape")
print(df.shape)
```

```python
# Data Quality Report (Before Cleaning)

print("\nBEFORE CLEANING")
print("total Rows",len(df))

print("total missing values")
df.isnull().sum()
```

```python
print("Duplicate Values:")
df.duplicated().sum()
```

```python
#Standarized Column Names
df.columns = df.columns.str.strip()
```

```python
# Handle Missing Values
# Fill Age with Median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill Embarked with Mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Fill Cabin with "Unknown"
df["Cabin"] = df["Cabin"].fillna("Unknown")
```

```python
print("total missing values")
df.isnull().sum()
```

```python
# Data Quality Report (After Cleaning)
# -----------------------------
print("\n========== AFTER CLEANING ==========")

print("Total Rows :", len(df))

print("\nMissing Values")
print(df.isnull().sum())

print("\nDuplicate Rows :", df.duplicated().sum())

# -----------------------------
# Generate Automated Cleaning Report
# -----------------------------
report = f"""
DATA CLEANING REPORT

Total Records : {len(df)}

Missing Values

{df.isnull().sum()}

Duplicate Records : {df.duplicated().sum()}

Cleaning Completed Successfully.
"""

with open("Cleaning_Report.txt", "w") as file:
    file.write(report)

print("\nCleaning Report Generated Successfully")

# -----------------------------
# Export Cleaned Dataset
# -----------------------------
df.to_csv("Titanic_Cleaned.csv", index=False)

print("Cleaned Dataset Saved Successfully!")
```
